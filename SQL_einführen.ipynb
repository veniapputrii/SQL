```python
#12. Feb 2026
#lerne über Datenbank in SQL

import sqlite3
import pandas as pd

#Ich öffne eine Verbindung zu einer neuen Datenbank

conn = sqlite3.connect('firma.db')

#Erstelle Test-Daten (Mitararbeiter)
daten = {
    'Name' : ['Anna', 'Nia', 'Clara', 'David'],
    'Abteilung' : ['IT', 'Sales', 'IT', 'HR'],
    'Gehalt' : [50000, 32000, 48000, 2900]
}
df_tmp = pd.DataFrame(daten)
#schreibe die Daten in die SQL-Datenbank
#die Tabelle in der Datenbank soll 'Mitarbeiter' heißen
df_tmp.to_sql('Mitarbeiter', conn, index=False, if_exists='replace')

print("Datenbank 'firma.db' erstellt und Tabelle 'Mitarbeiter' gefüllt !")
```

```python
#13. Feb. 2026
#Wir haben eine Tabelle für einen Online-Shop namens Kunden.
import sqlite3
import pandas as pd

#Verbindung zur Datenbank herstellen (produce)
conn = sqlite3.connect(':memory:')

#Tabelle erstellen
conn.execute('''
CREATE TABLE Kunden(
  ID INTEGER PRIMARY KEY,
  Name TEXT,
  Stadt TEXT,
  "Alter" INTEGER
)
''')

#Daten einfügen
kunden_daten= [
    (1, 'Max', 'Berlin', 30),
    (2, 'Nana', 'Hamburg', 28),
    (3, 'Lukas', 'Aachen', 32),
    (4, 'Julia', 'München', 29)
]
conn.executemany('INSERT INTO Kunden VALUES(?,?,?,?)', kunden_daten)

#Funktion zum Anzeigen der Ergebnisse
def sql(query):
  return pd.read_sql_query(query, conn)

print("Datenbank bereit!")
```

```python
sql("SELECT * FROM Kunden")
```
